## 3.Data prep for regression analysis

```python
# Create a scatterplot using matplotlib for another look at how chosen variables plot against each other.

Covid_country.plot(x= 'Total Tests', y = 'Active Cases', style ='o') # The style option creates a scatterplot; without it, we only have lines.
plt.title('Active Cases by Total Tests')
plt.xlabel('Total Tests')
plt.ylabel('Active Cases')
plt.show()
```

```python
# Reshape the variables into Numpy arrays and put them into separate objects.
x = Covid_country['Total Tests'].values.reshape(-1,1)
y = Covid_country['Active Cases'].values.reshape(-1,1)
```

```python
x 
```

```python
y
```

```python
# split data into training set and a test set.
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.3, random_state=0)
```

## Regression analysis

```python
# Create a regression object
# This is the regression object, which will be fit onto the training set.
regression = LinearRegression()
```

```python
# Fit the regression object onto the training set.
regression.fit(x_train, y_train)
```

```python
# Predict the values of y using x. 
y_predicted = regression.predict(x_test)
```

```python
# Create a plot that shows the regression line from the model on the test set.

plot_test = plt
plot_test.scatter (x_test, y_test, color='blue', s =15)
plot_test.plot(x_test, y_predicted, color ='red', linewidth =3)
plot_test.title('Active cases from the total tested(Test set)')
plot_test.xlabel('Total Tests')
plot_test.ylabel('Active Cases')
plot_test.show()
```

```python
# Create objects that contain the model summary statistics
# This is the mean squared error
rmse = mean_squared_error(y_test, y_predicted)
# This is the R2 score
r2 = r2_score(y_test, y_predicted)
```

```python
# Print the model summary statistics. This is where you evaluate the performnace of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```
