# Cleaning the data

```python
# importing the libraries
from pandas import DataFrame
import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Create path 
path = r'/Users/monajafizada/Documents/DataAnalytics/A6 advanced analytics and dashboard design/Project Information'
```

```python
# exporting the data of US deaths and Total tests 
C19_USA_City = pd.read_csv(os.path.join(path, '02 Data', 'PreparedData', 'covid_19_usa_city.csv'), index_col = False)
```

```python
# Checking the columns
C19_USA_City.columns
```

```python
# Dropping the extra columns
C19_USA_City = C19_USA_City.drop(columns=['Unnamed: 0','Date','New Deaths','Tests /1M pop','Deaths /1M pop','Total Cases /1M pop'])
```

```python
C19_USA_City.head()
```

```python
C19_USA_City.dtypes
```

```python
df = C19_USA_City
```

```python
df.shape
```

```python
# group by the states 
df = df.groupby(['City']).sum()
```

```python
df.head(100)
```

```python
df.shape
```

```python
# display all the rows from data frame
pd.set_option('display.max_rows', None)
```

```python
# make ascending by the total tests 
df_1 = DataFrame(df)
```

```python
test = df_1.sort_values(['Total Tests'], ascending =[0])
```

```python
test
```

### Exploratory Analysis

```python
# Creating scatterplot using matplot, ploting variables against each other
# the first option creates a scatterplot; without it, we only lines
test.plot(x= 'Total Tests',y='Total Deaths', style ='o')
plt.title('Covid death by the number of tests')
plt.xlabel('Total Tests')
plt.ylabel('Total Deaths')
plt.show
```

```python
# Reshape the variables into numpy arrays and put them into separate objects
x = test['Total Tests'].values.reshape(-1,1)
y = test['Total Deaths'].values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
# split data into a training set and a test set.
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

### Regression analysis

```python
# Create a regression object
regression = LinearRegression()
```

```python
# fit the regression object onto the training set
regression.fit(x_train, y_train)
```

```python
# predict the values of y using x
y_predicted = regression.predict(x_test)
```

```python
# Create a plot that shows the regression line from model on the test set
plot_test = plt
plot_test.scatter(x_train, y_train, color='blue', s=15)
plot_test.plot(x_test, y_predicted, color ='red', linewidth =3)
plot_test.title('COVID-19 Deaths by total test')
plot_test.xlabel('Total Tests')
plot_test.ylabel('Total Deaths')
plot_test.show()
```
