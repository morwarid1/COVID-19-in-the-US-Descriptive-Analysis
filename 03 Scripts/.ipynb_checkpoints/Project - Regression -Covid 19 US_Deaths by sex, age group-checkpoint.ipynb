### The elbow Technique 

```python
df =df[['Age','COVID-19 Deaths']]
```

```python
# defining the range of potential clusters in the data
num_cl = range(1,10)
# defining k-means clusters in the range assigned above
kmeans = [KMeans(n_clusters =i) for i in num_cl]
```

```python
# Createing a score that represents a rate of variation for the given cluster option
score =[kmeans[i].fit(df).score(df) for i in range (len(kmeans))]
score
# all the variables show be in flaot type
```

```python
# ploting the elbow curve using pylab
pl.plot(num_cl, score)
pl.xlabel('Number of clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

### K-means clustering

```python
# Create the k-means object
kmeans =KMeans(n_clusters = 3,n_jobs = -1)
```

```python
# fiting the k-means object to the data
kmeans.fit(df)
```

```python
# Create new column in the dataframe
df['clusters'] = kmeans.fit_predict(df)
```

```python
df.head()
```

```python
# how many observation fall into each clusters
df['clusters'].value_counts()
```

```python
# plot the clusters for the 'COVID-19 Deaths' and 'Age'
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['Age'], y=df['COVID-19 Deaths'], hue=kmeans.labels_, s=100)
ax.grid(False)
plt.xlabel('Age')
plt.ylabel('COVID-19 Deaths')
plt.show()
```
