```python
# importing you libraries and data
import quandl
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
import statsmodels.api as sm
import os
import warnings
warnings.filterwarnings ("ignore")
plt.style.use('fivethirtyeight')
import sklearn
from sklearn.cluster import KMeans
import pylab as pl
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Importing the data
dfM = pd.read_csv(os.path.join(path, '02 Data', 'PreparedData', 'ccdcmerged.csv'), index_col = False)
```

```python
# selecting only the necessary columns
dfM = dfM[["Confirmed","Deaths"]]
```

### Data prep for regression analysis

```python
# Ploting the data again with appropriate names
dfM.plot(x = 'Confirmed', y='Deaths',style='o') # The style option creates a scatterplot; without it, we only have lines.
plt.title('Covid-19 Deaths by confirmed cases')  
plt.xlabel('Confirmed')  
plt.ylabel('Deaths')  
plt.show()
```

```python
# reshaping the values 
X = dfM['Confirmed'].values.reshape(-1,1)
y = dfM['Deaths'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Split data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state =0)
```

### Regression analysis

```python
# setting up for the regression analysis
regression = LinearRegression ()
```

```python
regression.fit(X_train, y_train)
```

```python
y_predicted = regression.predict(X_test)
```

```python
# plot the clusters 
plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Covid 19 deaths cases by confirmed cases')
plot_test.xlabel('Confirmed')
plot_test.ylabel('Deaths')
plot_test.show()
```

### Regression Analysis

```python
# Create a scatterplot using matplotlib 
dfM.plot(x='Confirmed', y='Deaths')
plt.title('Confirmed cases over Deaths')
plt.xlabel('Confirmed Cases')
plt.ylabel('Death cases')
plt.show()
```

### The elbow Technigue

```python
num_cl = range(1,10)
kmeans = [KMeans(n_clusters =i) for i in num_cl]
```

```python
score = [kmeans[i].fit(dfM).score(dfM) for i in range (len(kmeans))]
score
```

```python
pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('score')
pl.title('Elbow Curve')
pl.show()
```

### Clustering Analysis

```python
# Create the K-means object 
kmeans = KMeans(n_clusters = 3, n_jobs = -1)
```

```python
#Fit the k _mean object to the data
kmeans.fit(dfM)
```

```python
dfM['clusters']= kmeans.fit_predict(dfM)
```

```python
dfM = dfM.reset_index()
```

```python
dfM['clusters'].value_counts()
```

```python
dfM.dtypes
```

```python
# plot the clusters 
plt.figure(figsize=(12,8))
ax =sns.scatterplot(x=dfM ['Confirmed'],y=dfM['Deaths'],hue =kmeans.labels_, s=100)
plt.xlabel('Confirmed')
plt.ylabel('Deaths')
plt.title('Covid-19 Confirmed Cases and Deaths Cases')
plt.show()  
```
