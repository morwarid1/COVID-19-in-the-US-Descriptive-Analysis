```python
dfD = dfD[["Province_State","2020-01-31","2020-02-29","2020-03-31","2020-04-30","2020-05-31","2020-06-30","2020-07-31","2020-08-31","2020-09-23"]]
```

```python
dfD = dfD.rename(columns={"Province_State":"State","2020-01-31":"Jan","2020-02-29":"Feb","2020-03-31":"Mar","2020-04-30":"Apr","2020-05-31":"May","2020-06-30":"Jun","2020-07-31":"Jul","2020-08-31":"Aug","2020-09-23":"Sept"})
```

### Mergin two columns

```python
# merging two Death and Confirmed data
dfM = pd.merge(dfC,dfD, on=["Months"])
```

```python
dfM.head(9)
```

```python
# export data
dfM.to_csv(os.path.join(path,'02 Data','PreparedData','deathsANDconfirmedBYmonths.csv'))
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = dfM["Months"]
y1 = dfM["Confirmed"]
y2 = dfM["Deaths"]

plt.figure(figsize=(20,10000000))
fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.ticklabel_format(style='plain')
ax.plot(x, y1, 'g-')
ax2.plot(x, y2, 'r')
ax.set_title('Covid-19 Deaths and Confirmed Cases by Months of 2020')
ax.set_xlabel('2020 Months')
ax.set_ylabel('Confirmed', color='g')
ax2.set_ylabel('Deaths', color='r')

plt.show()

# fig, ax1 = plt.subplots(x, y1, 'g-')
# ax2 = ax1.twinx()
# plt.figure(figsize=(12,8))
# plt.ticklabel_format(style='plain')
# ax1.plot(x, y1, 'g-')
# ax2.plot(x, y2, 'r-')

# plt.plot(x, y2, 'r-')
```
