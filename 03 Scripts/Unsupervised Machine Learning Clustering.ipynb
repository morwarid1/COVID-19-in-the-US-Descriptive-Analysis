#### 2. The elbow technique

```python
 num_cl = range(1, 10) # Defines the range of potential clusters in the data.
kmeans = [KMeans(n_clusters=i) for i in num_cl] # Defines k-means clusters in the range assigned above.
```

```python
score =[kmeans[i].fit(Covid_country).score(Covid_country) for i in range(len(kmeans))]
score 
```

```python
# Plot the elbow curve using Pylab

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

### 3. K-means clustering 

```python
# create the k-means object
kmeans =KMeans(n_clusters = 3, n_jobs = -1)
```

```python
# fit the K-means object to the data
kmeans.fit(Covid_country)
```

```python
Covid_country['clusters'] = kmeans.fit_predict(Covid_country)
```

```python
Covid_country.head()
```

```python
Covid_country['clusters'].value_counts()
```

```python
# plot the clusters for the "Total Tests" and " Active Cases" variables

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=Covid_country['Total Cases'], y=Covid_country['New Deaths'], hue=kmeans.labels_, s=100)
# Here, you are subsetting x for the x and y arguments to avoid using their labels.
# hue takes the value of the attribite kmeans.labels_
# which is the result of running the k-means algorithm
# s represents the size of the points you want to see in the plot.

ax.grid(False) # This rempves the grid from the background
plt.xlabel('Total Cases') # label x_axis
plt.ylabel('New Deaths') # label y_axis
plt.show()
```

```python
# plot the clusters for the "Total Tests" and " Active Cases" variables

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=Covid_country['Serious/Critical'], y=Covid_country['New Deaths'], hue=kmeans.labels_, s=100)
# Here, you are subsetting x for the x and y arguments to avoid using their labels.
# hue takes the value of the attribite kmeans.labels_
# which is the result of running the k-means algorithm
# s represents the size of the points you want to see in the plot.

ax.grid(False) # This rempves the grid from the background
plt.xlabel('Serious/Critical') # label x_axis
plt.ylabel('New Deaths') # label y_axis
plt.show()
```

```python
Covid_country.loc[Covid_country['clusters']==2, 'cluster'] = 'dark purple'
Covid_country.loc[Covid_country['clusters']==1, 'cluster'] = 'purple'
Covid_country.loc[Covid_country['clusters']==0, 'cluster'] = 'pink'
```
